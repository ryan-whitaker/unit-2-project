#### Calculate the average star rating for movies 2 hours or longer, and compare that with the average star rating for movies shorter than 2 hours.

```python
# Answer:
#We can see here that movies with a duration longer than 2 hours score a little hire on average than those below
print(movies[movies['duration']>120]['star_rating'].mean())
print(movies[movies['duration']<120]['star_rating'].mean())
```

```python
#Just for fun, I wanted to see what the total counts are for movies over and under 2 hours
print(movies[movies['duration']>120]['star_rating'].value_counts().sum())
print(movies[movies['duration']<120]['star_rating'].value_counts().sum())
```

#### Option 1: manually create a list of relevant genres, then filter using that list

```python
# Answer:
#Getting value counts for genres
movies['genre'].value_counts()
```

```python
#Creating new variable of genres with more than 10 counts
genres = list(movies['genre'].value_counts()[movies['genre'].value_counts() > 10].index)
```

```python
#Applying function to isolate movies in the genres variable and group them by genre and their respective average star rating
movies[movies['genre'].apply(lambda g: g in genres)].groupby('genre')['star_rating'].mean()
```
